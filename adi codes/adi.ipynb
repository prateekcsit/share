```python
'''
MODES:
Read- r | r+ | rb | rb+ | pointer at begining | data remain as it is
(if file does not exist, throws FileNotFoundError)
Write- w | w+ | wb | wb+ | pointer at begining | existing data truncated
(if file does not exist, creates new file) 
Append- a | a+ | ab | ab+ | pointer at end | data remain as it is
(if file does not exist, creates new file)
'''
```

### OOPS

```python
class Student:
    x=100
    
    def __init__(self, y):
        self.y=y
        
    def showxy(self):
        print(f'X={self.x} and Y={self.y}')
        return None
    
    def __del__(self):
        print('Object Deleted!')
```

```python
obj1=Student(200)
```

```python
obj1.x
```

```python
obj1.y
```

```python
obj1.showxy()
```

```python
obj2=Student(300)
```

```python
obj2.x
```

```python
obj2.y
```

```python
obj2.showxy()
```

```python
del obj2
```

```python
# Inheritance
class A:
    x=100
    def __init__(self, x=200):
        self.x=x
        
class B(A):
    y=300
    def __init__(self, y=500):
        # super().__init__()
        self.y=y
```

```python
objA=A()
objB=B()
```

```python
objA.x
```

```python
objB.y
```

```python
objB.x
```
